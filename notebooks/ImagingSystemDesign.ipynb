```python
# Imaging system design demo

import numpy as np
import matplotlib.pyplot as plt

# Simple thin lens formula
def image_position(f, do):
    return 1/(1/f - 1/do)

f = 0.05  # 5cm lens
object_distances = np.linspace(0.06, 0.5, 200)
images = image_position(f, object_distances)




plt.figure(figsize=(8,4))
plt.plot(object_distances, images)
plt.xlabel("Object Distance (m)")
plt.ylabel("Image Distance (m)")
plt.title("Imaging with a Thin Lens")
plt.grid(True)
plt.savefig("output_plots/imaging_system.png")
plt.show()
```
