```python
# Fresnel diffraction demo

import sys, os
sys.path.append(os.path.abspath('../'))

import numpy as np
import matplotlib.pyplot as plt
from core.waveoptics import fresnel_diffraction_1d

wavelength = 632.8e-9
x = np.linspace(-0.001, 0.001, 1000)
z = 0.5
aperture = lambda x: np.where(np.abs(x) < 0.0002, 1, 0)  # 0.4mm slit

U = fresnel_diffraction_1d(x, z, wavelength, aperture)
I = np.abs(U)**2




plt.figure(figsize=(8,4))
plt.plot(x*1e3, I/I.max())
plt.xlabel("x (mm)")
plt.ylabel("Normalized Intensity")
plt.title("Fresnel Diffraction Pattern at z=0.5m")
plt.grid(True)
plt.savefig("output_plots/fresnel_diffraction.png")
plt.show()
```
